# Data Preparation for Citibike Tableau Visualization

Note: All data from 2018

```python
# Add dependecies 
import glob
import pandas as pd
```

```python
df1 = pd.read_csv('raw_data\citibike_trip_histories2/201801-citibike-tripdata.csv')
df1.head()
```

```python
df1.rename(columns = {'Trip Duration': 'tripduration', 'Start Time': 'starttime', 
                      'Stop Time': 'stoptime', 'Gender':'gender', "User Type": "usertype",
                      'Bike ID': 'bikeid', 'Birth Year': 'birth year', 'Start Station ID': 'start station id',
                      'Start Station Name': 'start station name', 'Start Station Latitude': 'start station latitude',
                      'Start Station Longitude': 'start station longitude', 'End Station ID': 'end station id', 
                      'End Station Name': 'end station name', 'End Station Latitude': 'end station latitude',
                      'End Station Longitude': 'end station longitude'}, inplace = True)
df1.head()
```

```python
df2 = pd.read_csv('raw_data\citibike_trip_histories2/201802-citibike-tripdata.csv')
df2.head()
```

```python
df2.rename(columns = {'Trip Duration': 'tripduration', 'Start Time': 'starttime', 
                      'Stop Time': 'stoptime', 'Gender':'gender', "User Type": "usertype",
                      'Bike ID': 'bikeid', 'Birth Year': 'birth year', 'Start Station ID': 'start station id',
                      'Start Station Name': 'start station name', 'Start Station Latitude': 'start station latitude',
                      'Start Station Longitude': 'start station longitude', 'End Station ID': 'end station id', 
                      'End Station Name': 'end station name', 'End Station Latitude': 'end station latitude',
                      'End Station Longitude': 'end station longitude'}, inplace = True)
df2.head()
```

```python
df3 = pd.read_csv('raw_data\citibike_trip_histories2/201803-citibike-tripdata.csv')
df3.head()
```

```python
df3.rename(columns = {'Trip Duration': 'tripduration', 'Start Time': 'starttime', 
                      'Stop Time': 'stoptime', 'Gender':'gender', "User Type": "usertype",
                      'Bike ID': 'bikeid', 'Birth Year': 'birth year', 'Start Station ID': 'start station id',
                      'Start Station Name': 'start station name', 'Start Station Latitude': 'start station latitude',
                      'Start Station Longitude': 'start station longitude', 'End Station ID': 'end station id', 
                      'End Station Name': 'end station name', 'End Station Latitude': 'end station latitude',
                      'End Station Longitude': 'end station longitude'}, inplace = True)
df3.head()
```

```python
citibike_2017q1 = pd.concat([df1, df2, df3])
citibike_2017q1.head()
```

```python
citibike_2017q1.to_csv('raw_data/citibike_trip_histories/citibike_2017q1.csv')
```

```python
# glob.glob('data*.csv') - returns List[str]
# pd.read_csv(f) - returns pd.DataFrame()
# for f in glob.glob() - returns a List[DataFrames]
# pd.concat() - returns one pd.DataFrame()

ridership_2018 = pd.concat([pd.read_csv(f) for f in glob.glob('raw_data/citibike_daily_ridership_membership_data/*.csv')],\
                                 ignore_index = True)
trip_2018 = pd.concat([pd.read_csv(f) for f in glob.glob('raw_data/citibike_trip_histories/*.csv')],\
                            ignore_index = True)
```
